```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r'/home/rahul/WSL_Projects/ReferencesPython/Convolutional Neural Networks/Image_Processing/Wolf.jpeg')
image1 = image.copy()
gray = cv2.cvtColor(image1,cv2.COLOR_RGB2GRAY)
blur = cv2.GaussianBlur(gray,(5,5),0)
thresh = cv2.threshold(blur,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)[1] 
#convert image into Binary, OTSU is used to seperate foreground and background
counts=cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

counts=counts[0] if len(counts)==2 else counts[1]
for c in counts:
    area= cv2.contourArea(c)
    if area>10000:
        cv2.drawContours(image1,[c],-1,(36,255,12),3)

# cv2.imwrite('thresh.png',thresh)
# cv2.imwrite('image.png',image1)
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize = (10,15), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(image)
ax1.set_title('Input Image')

ax2.axis('off')
ax2.imshow(image1)
ax2.set_title('Objects Detected')
```
