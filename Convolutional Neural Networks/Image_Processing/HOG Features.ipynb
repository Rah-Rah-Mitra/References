```python
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data,exposure
```

```python
image = data.astronaut()
fd , hog_image = hog(image, orientations=8, pixels_per_cell=(16,16), cells_per_block=(1,1), visualize=True, channel_axis=-1)
# orientation is angle, pixel per cell is kernel size, cell per block is the comparision block that calculates the filter, visualize just tells us if we need visualisation

fig, (ax1, ax2) = plt.subplots(1,2, figsize = (8,4), sharex = True, sharey = True)

ax1.axis('off')
ax1.imshow(image, cmap= plt.cm.gray)
ax1.set_title('input image')

hog_image_rescale = exposure.rescale_intensity(hog_image, in_range=(0,10)) #rescale needed for better display

ax2.axis('off')
ax2.imshow(hog_image_rescale, cmap= plt.cm.gray)
ax2.set_title('input imageHistogram of Gradients')

plt.show()
```
