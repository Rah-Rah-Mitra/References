```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img= cv2.imread(r"/home/rahul/WSL_Projects/ReferencesPython/Convolutional Neural Networks/Image_Processing/bird.png")
h,w=img.shape[0:2]

base_size=h+20, w+20, 3 #padding of 20 pixels

base= np.zeros(base_size,dtype=np.uint8)
cv2.rectangle(base,(0,0), (w+20,h+20),(0,0,0),30)
base[10:h+10,10:w+10] = img
```

```python
base= cv2.cvtColor(base,cv2.COLOR_BGR2RGB)
img= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

fig, (ax1, ax2) = plt.subplots(1,2, figsize = (8,4), sharex = True, sharey = True)

ax1.axis('off')
ax1.imshow(img, cmap= plt.cm.gray)
ax1.set_title('Input image')

ax2.axis('off')
ax2.imshow(base, cmap= plt.cm.gray)
ax2.set_title('Padded Image')

plt.show()
```
