MxNet also sypports symbolic math. This is more like the algebra you learned in school, where you can encode a formula, and the variables can take on different values.

Symbolic math allows MxNet to know the shape of your formula, compute given specific values, and automatically compute gradients. This can be used as the 'raw' interface to create neural networks, but don't worry -- this relatively math heavy approach isn't required -- we'll cover a much more engineer friendly interface in Gluon!

```python
import mxnet as mx
```

Here we are making a relatively simple formula. This does seem like a fairly verbose way to say `2 * a + b`, but what we are really doing is making a small symbolic meta program that we can vary the values.

```python
a = mx.sym.Variable('a')
b = mx.sym.Variable('b')
c = 2 * a + b
```

Now we can bind variables, think of this as a function call in regular programming.

```python
e = c.bind(mx.cpu(), {'a': mx.nd.array([3]), 'b':mx.nd.array([2])})
```

And run the formula - which is really a symbolic math *network* in the forward direction and get a numerical value.

```python
e.forward()
```

```python
mx.viz.plot_network(symbol=c)
```
