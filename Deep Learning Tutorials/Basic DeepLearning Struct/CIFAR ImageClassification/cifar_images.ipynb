```python
import keras
from keras.datasets import cifar10
import matplotlib.pyplot as plt
import numpy as np
```

```python
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

Taking a look at one of the training images. The 10 different classes represent airplanes, cars, birds, cats, deer, dogs, frogs, horses, ships, and trucks.

```python
fig = plt.figure()
for i in range(0, 10):
    ax = fig.add_subplot(1, 10, i+1)
    ax.imshow(train_images[i])
    plt.axis('off')
train_images.shape
```
